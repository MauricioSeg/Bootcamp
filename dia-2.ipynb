```python
temperatura_usuario = int(input("Que temperatura desea?"))
print (temperatura_usuario,"°c")
def temperatura_del_Aire(temperatura): 
    if temperatura >= -10 and temperatura <= 13:
        print("Ro'y koko")
    elif temperatura >= 14 and temperatura <= 23:
        print("Refresca'i")
    elif temperatura >= 24 and temperatura <= 28:
        print("Oikoite")
    elif temperatura >= 29 and temperatura <= 45:
        print("Hakurooooo")
    else:
        print("Ñamano mba'e!")
temperatura_del_Aire(temperatura_usuario)
```
